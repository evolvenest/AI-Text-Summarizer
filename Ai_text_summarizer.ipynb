```python
import gradio as gr

# Function to summarize text
def summarize_text(user_text):
    prompt = f"""
    Summarize the following text in 3 concise bullet points:

    {user_text}
    """

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )

    return response.text

# Create Gradio Interface
demo = gr.Interface(
    fn=summarize_text,
    inputs=gr.Textbox(
        lines=10,
        placeholder="Enter text to summarize...",
        label="Input Text"
    ),
    outputs=gr.Textbox(
        lines=8,
        label="Summary"
    ),
    title="AI Text Summarizer",
    description="Enter long text and get a concise summary using Gemini AI."
)

# Launch the app
demo.launch()
```
